```python
# Install NLTK (if not already installed)
!pip install nltk scikit-learn --quiet

# Import libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk

# Download tokenizer models
nltk.download('punkt')

# Sample corpus of 10 documents
documents = [
    "Natural Language Processing is a fascinating field.",
    "TF-IDF stands for Term Frequency-Inverse Document Frequency.",
    "Stemming reduces words to their base form.",
    "Lemmatization also reduces words but with context.",
    "Ambiguity in language makes NLP challenging.",
    "TF and IDF are the components of the TF-IDF weighting.",
    "A bank can refer to a financial institution or a river bank.",
    "Running and ran are forms of the word run.",
    "Better is the lemmatized form of good in some contexts.",
    "Lexical ambiguity involves word meanings."
]

# Step 1: Create a TF-IDF vectorizer and fit to documents
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)

# Step 2: Convert the matrix to a DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Step 3: View the top 10 rows and all columns
pd.set_option('display.max_columns', None)
display(tfidf_df.head(10))
```

```python
# Install packages
!pip install PyMuPDF scikit-learn pandas regex --quiet

import fitz  # PyMuPDF
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import re

# Upload PDF
from google.colab import files
uploaded = files.upload()
pdf_file = next(iter(uploaded))

# Extract text
doc = fitz.open(pdf_file)
pdf_text = " ".join([page.get_text() for page in doc])

# Clean text: remove numbers and symbols
def clean_text(text):
    text = re.sub(r'\d+', '', text)  # remove numbers
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    return text.lower()

cleaned_text = clean_text(pdf_text)

# Chunk into small paragraphs (like documents)
corpus = [para.strip() for para in cleaned_text.split('\n') if len(para.strip()) > 30]

# Vectorize TF-IDF
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(corpus)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out())

# Display top 10 rows
tfidf_df.head(10)
```

```python
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import PorterStemmer
import pandas as pd

tokenizer = TreebankWordTokenizer()
stemmer = PorterStemmer()

# Use cleaned_text from Step 1
sentences = [sent.strip() for sent in cleaned_text.split('.') if len(sent.strip()) > 20]

original = []
stemmed = []

for sent in sentences[:10]:  # Show first 10
    tokens = tokenizer.tokenize(sent)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    original.append(" ".join(tokens))
    stemmed.append(" ".join(stemmed_tokens))

# Display results
df_stemmed = pd.DataFrame({
    'Original Sentence': original,
    'Stemmed Sentence': stemmed
})
df_stemmed
```

```python
# Install spaCy and English model
!pip install -U spacy --quiet
!python -m spacy download en_core_web_sm

import spacy
import pandas as pd

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Split text into sentences
sentences = [s.strip() for s in cleaned_text.split('.') if len(s.strip()) > 20]

original = []
lemmatized = []

# Lemmatize top 10 sentences
for sent in sentences[:10]:
    doc = nlp(sent)
    original.append(" ".join([token.text for token in doc]))
    lemmatized.append(" ".join([token.lemma_ for token in doc]))

# Show comparison
df_lemma_spacy = pd.DataFrame({
    "Original Sentence": original,
    "Lemmatized Sentence": lemmatized
})
df_lemma_spacy

```

```python
import spacy
from spacy import displacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Example sentence with ambiguity
ambiguous_sentence = "I saw the man with the telescope."

# Process the sentence
doc = nlp(ambiguous_sentence)

# Display part-of-speech and dependency parse
for token in doc:
    print(f"{token.text:10s} | POS: {token.pos_:8s} | Dep: {token.dep_:10s} | Head: {token.head.text}")

# Visualize the dependency tree (in Colab)
displacy.render(doc, style='dep', jupyter=True)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

text = "He went to the bank to withdraw money. The fisherman sat on the bank of the river."
doc = nlp(text)

for token in doc:
    if token.text.lower() == "bank":
        print(f"{token.text:5s} | POS: {token.pos_:8s} | Context: {token.sent.text}")
```

🧠 This shows how "bank" has the same spelling but changes meaning depending on context.

```python
text = "I saw the man with the telescope."
doc = nlp(text)

for token in doc:
    print(f"{token.text:10s} | Dep: {token.dep_:12s} | Head: {token.head.text}")

# Visualize
from spacy import displacy
displacy.render(doc, style="dep", jupyter=True)
```

👀 This shows multiple possible parse trees for one sentence.


```python
text = "Can you pass the salt?"
doc = nlp(text)
for token in doc:
    print(f"{token.text:10s} | POS: {token.pos_:8s} | Dep: {token.dep_:12s}")
```
